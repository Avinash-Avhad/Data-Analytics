# Exploratory Data Analysis on "Global Stock Exchanges (Cap = 1 trillion+) 04-23" Dataset

# Year

```python
plt.figure(figsize=(15,5))
a=sns.countplot(data=df,x='Year')
plt.title("Value Count Of Year Column")
for b in a.containers:
    a.bar_label(b)
```

```python
ex=df.groupby(['Year'],as_index=False)['Value'].sum().sort_values(by='Value',ascending=False)
print(ex)
plt.figure(figsize=(15,5))
a=sns.barplot(x='Year',y='Value',data=ex)
plt.title("Year VS Value")

for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that The highest count Of The number of trade Was in 2021,2015.

but the highest value of trade was in 2021,2023,2020

```python
ex=df.groupby(['Year','Region'],as_index=False)['Value'].sum().sort_values(by='Value',ascending=False)
print(ex)
plt.figure(figsize=(15,5))
a=sns.barplot(x='Year',y='Value',data=ex,hue='Region',alpha=0.9)
plt.title("Year VS Value Categorise with Regions")
for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the Value of the trade was high in 2021,2023,2020 in America Region

```python
ex=df.groupby(['Year'],as_index=False)['YTD'].sum().sort_values(by='YTD',ascending=False)
print(ex)
plt.figure(figsize=(15,5))
sns.set_theme(style='ticks')
a=sns.barplot(x='Year',y='YTD',data=ex)
plt.title("Year VS YTD")
for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the the yearly performance of the stock market was high in 2021,2023,2020,2022

```python
ex=df.groupby(['Year','Region'],as_index=False)['YTD'].sum().sort_values(by='YTD',ascending=False)
print(ex)
plt.figure(figsize=(25,10))
sns.set_theme(style='white')
plt.title("Year VS YTD Classified With Distinct Region")
a= sns.barplot(data=ex,x='Year',y='YTD',hue='Region')
for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the the yearly performance of the stock market was high in 2021,2023,2020,2022

but region wise yearly performance was high in the America Region in the year of 2023,2021,2020

```python
ex=df.groupby(['Year','Region'],as_index=False)['% Change (YTD)'].sum().sort_values(by='% Change (YTD)',ascending=False)
print(ex)
plt.figure(figsize=(15,6))
sns.set_theme(style='whitegrid')
plt.title("Year VS % Change (YTD) Classified With Distinct Region")
a= sns.pointplot(data=ex,x='Year',y='% Change (YTD)',hue='Region')
for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the year wise % change in (YTD) Or the % change in Tearnover of stock market was high in 2021 in Europe - Africa-Middle East Region

# Region

```python
plt.figure(figsize=(10,5))
sns.set_theme(style='whitegrid')
plt.title("Region Wise Count")
a=sns.countplot(data=df,x='Region')

for b in a.containers:
    a.bar_label(b)
```

```python
region_counts = df['Region'].value_counts()
N=[0.7,0.1,0]
sns.set_theme(style='ticks')
plt.pie(region_counts, labels=region_counts.index, colors=[ 'gold', 'violet', 'pink'], autopct='%1.1f%%',pctdistance=0.5,
    rotatelabels=True,shadow=True,
    labeldistance=1.3,
    startangle=1,
    radius=1.5,explode=N)
plt.text(1.3,2.0,"The Asia - Pacific Have Highest Count In World Trade.\n That Is 42.1% Of Trade Done In Asia - Pacific Region ")
# Add title and show the plot
#plt.title("Distribution of Regions")
plt.show()
```

```python
ex=df.groupby(['Region'],as_index=False)['Value'].sum().sort_values(by='Value',ascending=False)
print(ex)
plt.figure(figsize=(10,5))
sns.set_theme(style='dark')
plt.title("Region VS Value")
a=sns.barplot(x='Region',y='Value',data=ex)

for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the count of Region Means The no of trade Transaction Was high in Asia - Specific Region.

from pie chart we can see that the The Asia - Pacific Has Highest Count In World Trade.That Is 42.1% Of Trade Done In only Asia - Pacific Region.

according to pie chart the america was only 16.9% count in world trade but according to value the America Region Value was Highest than other.

```python
ex=df.groupby(['Region'],as_index=False)['Nominal'].sum().sort_values(by='Nominal',ascending=False)
print(ex)
plt.figure(figsize=(10,5))
sns.set_theme(style='white')
plt.title("Region VS Nominal")
a=sns.barplot(x='Region',y='Nominal',data=ex,saturation=0.9)

for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the Nominal Amount Of Thade Also Was in America Region Means The Thrade Done In America Region Was in high Amount Compare to another Region.

```python
ex=df.groupby(['Region','ExchangeName'],as_index=False)['YTD'].sum().sort_values(by='YTD',ascending=False)
print(ex)
plt.figure(figsize=(15,5))
sns.set_theme(style='whitegrid')
plt.title("ExchangeName VS YTD Classified With Regions")
a=sns.barplot(y='ExchangeName',x='YTD',data=ex,hue='Region')

for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the Region Wise Yearly Performance (YTD) was high in America Region Dou to the NYSE & Nasdaq - US Stock Exchanger.

```python
ex=df.groupby(['Region','ExchangeName'],as_index=False)['% Change (YTD)'].sum().sort_values(by='% Change (YTD)',ascending=False)
print(ex)
plt.figure(figsize=(15,5))
sns.set_theme(style='darkgrid')
plt.title("ExchangeName VS % Change (YTD) Classified With Regions")
a=sns.barplot(y='ExchangeName',x='% Change (YTD)',data=ex,hue='Region')

for b in a.containers:
    a.bar_label(b)
```

from above graphs we can see that the year wise % change in (YTD) Or the % change in Tearnover of stock market was high in 2021 in Europe - Africa-Middle East Region
Due To the LSE Group London Stock Exchanger

```python
a=df[['Value','YTD','% Change (YTD)']].corr()
a
sns.set_theme(style='ticks')
plt.title("Correlation Between The Value,YTD,% Change(YTD)")
sns.heatmap(a,annot=True,linecolor='c',linewidths='06',square=True)
```

From Above Hitmap We Can See that The Relation ship Between The Value & YTD Has Good with each other But the both has Bad relation with %change(YTD)
Due to This According to value Change in America Region Also Increse The YTD But Not Increse in % change (YTD)

```python
a=[]
am=[]
e=[]
for i,j in zip(df['Region'],df['ExchangeName']):
    if i=='Americas':
        am.append(j)
    elif i=='Asia - Pacific':
        a.append(j)
    else:
        e.append(j)
    #print(i,j)
#Americas
#Asia - Pacific
#Europe - Africa - Middle East
print("American Stock Exchanger : \n",am)
print("---------------------------------------------------------------------------------------------------------\n")
print("Asian Stock Exchanger : \n",a)
print("---------------------------------------------------------------------------------------------------------\n")
print("Europe Stock Exchanger : \n",e)
```

```python
ex=df.groupby(['Year','ExchangeName'],as_index=False)['YTD'].sum().sort_values(by='YTD',ascending=False)
print(ex)
plt.figure(figsize=(15,25))
sns.set_theme(style='whitegrid')
plt.title("ExchangeName VS % Change (YTD) Classified With Regions")
sns.barplot(y='ExchangeName',x='YTD',data=ex,hue='Year')
```

From above Graph we can see that NYSC, Nasdaq - US Exchangers from America Region was large YTD Between 2020 To 2023 only and thise rate of YTD is not Constant For Every Year i.e the % Cange (YTD) is Decrese.

but from Europe - Africa-Middle East Region the LSE Group London Stock Exchanger Not have large YTD but for every year this YTD Rate are constant i.t. The % change (YTD) of LSE Group London Stock Exchanger is large
